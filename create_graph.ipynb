```python
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np
```

```python
def index_to_color(index):
    if (index ==0): 
        return "#a5f291"
    if (index ==1): 
        return "#4b38f5"
    if (index ==2): 
        return "#a6a1cf"
    if (index ==3): 
        return "#31d4cc"
    if (index ==4): 
        return "#9211cf"
    if (index ==5): 
        return "#10c43a"
    if (index ==6): 
        return "#e7eb28"
    if (index ==7): 
        return "#e37346"
    if (index ==8): 
        return "#b03200"
    if (index ==9): 
        return "#ba278e"
    if (index ==10): 
        return "#db91c5"
    if (index == 11):
        return "#0b03fc"
    if (index == 12):
        return "#0b03fc"
    if (index == 13):
        return "#d13bae"
    if (index == 14):  
        return "#cc1225"  
inputpath = "data/data_stability_init.csv"
df = pd.read_csv(inputpath, sep=",")

#df = df.T.reset_index()
#df.to_csv("datos_grafico_raw2_clean_traspose.csv", sep=';', index = False)

print(df)

new_simulated_rows = []
for index, row in df.iterrows():
    initial_diagnose = row['initial']
    print(row)
    for column in df:
        if(not column == "initial"):
            for i in range(0,int(row[column])):
                new_item = [initial_diagnose, column, index_to_color(index)]
                new_simulated_rows = new_simulated_rows + [new_item]


df_simulated = pd.DataFrame(np.array(new_simulated_rows), columns=['initial diagnose', 'final diagnose', 'color'])

print(df_simulated)

df_simulated.to_csv("data/datos_grafico_recreados.csv", sep=';', index = False)
```
